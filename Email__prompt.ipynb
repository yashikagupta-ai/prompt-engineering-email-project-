```python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

openai.api_key  = os.getenv('OPENAI_API_KEY')
```

```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
```

```python
# Project: Prompt Engineering - Professional Email Rewrite
# Platform: DeepLearning.AI
# Author: Yashika Gupta, Mahindra University
# Description: This project demonstrates prompt refinement by transforming a casual follow-up email into a professional tone, improving gratitude, and making it concise.

# Prompt 1: Transforming a casual email into formal tone
# Prompt 2: Improving tone to sound more grateful
# Prompt 3: Making the professional email more confident yet humble
```

```python
# This is a very casual email given by the user
original_email = """
Hey, just checking in. Any update on the interview?
"""
# Instruction to the model: Rewrite the casual email into a more professional tone,
#by imagining you're a hiring manager receiving it from a candidate

instruction = """You are a hiring manager at a company. A candidate has sent you a follow-up email after an interview, but it is very casual. \
Rewrite it in a professional, polite, and enthusiastic tone as you would prefer to receive."""

prompt = f"""
Original Email:
\"\"\"{original_email}\"\"\"

{instruction}
"""

response = get_completion(prompt)
print("‚úçÔ∏è Rewritten Email:\n")
print(response)
```

```python
rewritten_email = """
Subject: Follow-Up on Interview Status

Dear [Hiring Manager's Name],

I hope this message finds you well. I wanted to inquire about any updates regarding the recent interview I had with your team.\
I am very enthusiastic about the opportunity to potentially join [Company Name] and contribute to its success.

Thank you for considering my application, and I look forward to hearing from you soon.

Warm regards,

[Your Name]
"""
#Here new instruction is to change the tone to more grateful to the interviewer and the team
new_instruction = """
This is a follow-up email draft written after an interview.
Improve it by making the applicant sound more grateful for the interviewer's time and effort.
Keep it formal and sincere.
"""
```

```python
prompt2 = f"""
Email Draft:
\"\"\"{rewritten_email}\"\"\"

{new_instruction}
"""
response2 = get_completion(prompt2)
print("üì® Improved Email with Gratitude:\n")
print(response2)
```

```python
prompt3 = f"""
Email Draft:
\"\"\"{rewritten2_email}\"\"\"

{new_instruction2}
"""
response3 = get_completion(prompt3)
print("üì® Improved Email :\n")
print(response3)
```
